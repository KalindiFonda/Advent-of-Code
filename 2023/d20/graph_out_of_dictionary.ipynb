```python
# thank you internet for this code!

import matplotlib.pyplot as plt
import networkx as nx

# Your tree structure
tree_data = {} # REDACTED dictionary created from the nodes and it's connections
# Create a directed graph
G = nx.DiGraph()
plt.figure(figsize=(20, 20))

# Add nodes and edges to the graph
for node, neighbors in tree_data.items():
    G.add_node(node)
    G.add_edges_from((node, neighbor) for neighbor in neighbors)

# Draw the graph
nx.draw(G, font_weight='bold', node_size=700, node_color='skyblue', arrowsize=15)

# Show the plot
plt.show()
```
